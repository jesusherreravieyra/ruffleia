```python
import pygame
from random import randint
import textwrap

ANCHO = 974
ALTO = 620

BLANCO = (255, 255, 255)
VERDE_BANDERA = (27, 94, 32)
ROJO = (255, 0, 0)
AZUL = (0, 0, 255)
NEGRO = (0, 0, 0)

MENU = 1
JUGANDO = 2
GANAR=3
PERDER =4
QUITAR=5

QUIETO = 1
ABAJO = 2
ARRIBA = 3
IZQUIERDA = 4
DERECHA =5

APARECE = 2
INVISIBLE = 1
VIAJANDO=3

estadoBalon = INVISIBLE
estado = MENU

factos = ['Pepsi in Space: In 1985, Pepsi became the first soft drink company to launch a soda can into space aboard the Space Shuttle Challenger. This marketing stunt was part of Pepsi´s Pepsi Challenge campaign', 
          'Origins in Pharmacies: Pepsi was first created in 1893 by Caleb Bradham, a pharmacist from North Carolina. Originally called "Brad´s Drink," it was later renamed Pepsi-Cola in 1898 after the digestive enzyme pepsin and kola nuts used in the recipe.',
          'Do Us a Flavor Contest: Lay´s runs the "Do Us a Flavor" contest where consumers can suggest new chip flavors. Winners have included unusual flavors like "Southern Biscuits and Gravy" and "Wasabi Ginger," which were later produced and sold nationally.',
          'World´s Largest Chip: In 2013, Lay´s set a Guinness World Record for creating the largest potato chip. The chip measured 25 by 14 inches and was presented at the International Food Exhibition in Paris.',
          'Flavor Innovation: Doritos are known for their wide range of bold flavors. Some iconic flavors include Nacho Cheese, Cool Ranch, Spicy Nacho, and more recently, Blaze and Flamin´ Hot.',
          'Super Bowl Advertising: Doritos has become synonymous with memorable Super Bowl commercials. The "Crash the Super Bowl" campaign invited consumers to create their own Doritos ads, with winners having their commercials aired during the Super Bowl, often becoming fan favorites.',
          'Variety of Shapes and Sizes: Cheetos come in various shapes and sizes, including Crunchy Cheetos, Puffs, and even specialized shapes tied to promotions like Cheetos Bag of Bones for Halloween.',
          'Cheetos Museum: In 2017, Cheetos launched an online museum called the Cheetos Museum where fans could submit uniquely shaped Cheetos they found. Winners received cash prizes and their Cheetos were featured in the virtual museum.']

ARCHIVO = open("settings.txt", "r+")
record = float(ARCHIVO.readlines()[len(ARCHIVO.readlines())-1])


def dibujarMenu(ventana, imgBtnJugar):
    fuente = pygame.font.SysFont("monospace", 30)
    ventana.blit(imgBtnJugar, (250, +400))


def dibujarBalones(ventana, listaBalones):
    for balon in listaBalones:
        ventana.blit(balon.image, balon.rect)


def moverBalones(listaBalones):
    for balones in listaBalones:
        balones.rect.left += 7


def dibujarPersonaje(ventana, spritePersonaje):
    ventana.blit(spritePersonaje.image, spritePersonaje.rect)


def dibujarEnemigos(ventana, listaEnemigos):
    for enemigo in listaEnemigos:
        ventana.blit(enemigo.image, enemigo.rect)


def verificarColision(listaEnemigos,listaBalones, listaTargets):
    global estadoBalon
    global estado
    efecto3 = pygame.mixer.Sound("golpe.wav")
    for k in range (len(listaBalones)-1,-1,-1):
        balon = listaBalones[k]
        for e in range(len(listaEnemigos)-1,-1,-1):
            enemigo = listaEnemigos[e]
            xe,ye,ae,alte = enemigo.rect
            xb,yb,ab,altb = balon.rect
            if xb>=xe and xb<=xe+ae and yb >= ye and yb <=ye+alte: #Le pegó
                efecto3.play()
                listaEnemigos.remove(enemigo)
                listaBalones.remove(balon)
                estadoBalon=INVISIBLE
                listaTargets.append(1)

                break


def verificarSlida (listaBalones):
    global estadoBalon
    for k in range (len(listaBalones)-1,-1,-1):
        balon = listaBalones[k]
        xb, yb, ab, altb = balon.rect
        ancho = ANCHO
        if xb>=ancho:
            listaBalones.remove(balon)
            estadoBalon = INVISIBLE
            break


def dibujar():
    global estadoBalon
    global estado
    pygame.init()
    ventana = pygame.display.set_mode((ANCHO, ALTO))
    reloj = pygame.time.Clock()
    termina = False

    #Targets destruidas
    listaTargets = []
    listaTargetsMalos = []

    # Personaje
    imgPersonaje = pygame.image.load('ruffilio.png')
    imgPersonaje = pygame.transform.scale(imgPersonaje, (100, 143.3))
    spritePersonaje = pygame.sprite.Sprite()
    spritePersonaje.image = imgPersonaje
    spritePersonaje.rect = imgPersonaje.get_rect()
    spritePersonaje.rect.left = 0

    listaEnemigos = []
    imgEnemigo = pygame.image.load("pepsi_zero.png")
    imgEnemigo= pygame.transform.scale(imgEnemigo, (60, 60))

    for k in range(randint(6,20)):
        spriteEnemigo = pygame.sprite.Sprite()
        spriteEnemigo.image = imgEnemigo
        spriteEnemigo.rect = imgEnemigo.get_rect()
        spriteEnemigo.rect.left = randint(200, ANCHO-100)
        spriteEnemigo.rect.bottom = randint(spriteEnemigo.rect.height+50, ALTO-10)
        listaEnemigos.append(spriteEnemigo)

    listaLays = []
    imgLays = pygame.image.load("lays.png")
    imgLays= pygame.transform.scale(imgLays, (60, 90))
    imgCheetos = pygame.image.load("Cheetos.png")
    imgCheetos= pygame.transform.scale(imgCheetos, (60, 90))
    imgdoritos = pygame.image.load("doritos.png")
    imgdoritos= pygame.transform.scale(imgdoritos, (60, 90))
    for k in range(randint(1,4)):
        spriteLays = pygame.sprite.Sprite()
        spriteLays.image = imgLays
        spriteLays.rect = imgLays.get_rect()
        spriteLays.rect.left = randint(200, ANCHO-100)
        spriteLays.rect.bottom = randint(spriteLays.rect.height+50, ALTO-10)
        listaLays.append(spriteLays)

        spriteLays = pygame.sprite.Sprite()
        spriteLays.image = imgCheetos
        spriteLays.rect = imgCheetos.get_rect()
        spriteLays.rect.left = randint(200, ANCHO-100)
        spriteLays.rect.bottom = randint(spriteLays.rect.height+50, ALTO-10)
        listaLays.append(spriteLays)

        spriteLays = pygame.sprite.Sprite()
        spriteLays.image = imgdoritos
        spriteLays.rect = imgdoritos.get_rect()
        spriteLays.rect.left = randint(200, ANCHO-100)
        spriteLays.rect.bottom = randint(spriteLays.rect.height+50, ALTO-10)
        listaLays.append(spriteLays)

    listaBalones = []
    imgBalon = pygame.image.load("balonF.png")
    imgBalon = pygame.transform.scale(imgBalon, (20.48, 20.48))


    #Button click to play
    imgBtnJugar = pygame.image.load("button.png")
    imgWIN =pygame.image.load("win.png")    
    imgWIN = pygame.transform.scale(imgWIN, (240, 240))
    imgLOSE =pygame.image.load("lose.png")    
    imgLOSE = pygame.transform.scale(imgLOSE, (260, 90))
    #Background images
    imgFondo = pygame.image.load ("canchaf.png")
    imgFondo = pygame.transform.scale(imgFondo, (943.5, 540))
    imgFondoMenu = pygame.image.load ("background.png")
    imgFondoMenu= pygame.transform.scale(imgFondoMenu, (ANCHO, ALTO))
    img2 = pygame.image.load("background.png")
    img2= pygame.transform.scale(img2, (ANCHO, ALTO))
    img3 = pygame.image.load("background.png")
    img3= pygame.transform.scale(img3, (ANCHO, ALTO))
    imgRuffle= pygame.image.load("ruffle.png")


    moviendo = QUIETO

    # Audio
    if estado == MENU:
        pygame.mixer.init()

    efecto = pygame.mixer.Sound ("SILBATO.wav")
    efecto2 = pygame.mixer.Sound ("patada.wav")
    efecto3 = pygame.mixer.Sound("aplausoss.wav")
    efecto4 = pygame.mixer.Sound("trompteta.wav")

    # Tiempo
    timer = 0
    fuente = pygame.font.SysFont("monospace", 30)
    fuente2 = pygame.font.SysFont("monospace", 60,True)
    fuente3 = pygame.font.SysFont("monospace",27,True)
    fuente4 = pygame.font.SysFont("monospace",30,True)
    fuente5 = pygame.font.SysFont("monospace", 20,True)
    fuente6 = pygame.font.SysFont("monospace", 30)

    while not termina:
        for evento in pygame.event.get():
            if evento.type == pygame.QUIT:
                termina = True
            elif evento.type == pygame.KEYUP:
                moviendo = QUIETO
            elif evento.type == pygame.KEYDOWN:
                if evento.key == pygame.K_UP:
                    moviendo = ARRIBA
                elif evento.key == pygame.K_DOWN:
                    moviendo=ABAJO
                elif evento.key == pygame.K_LEFT:
                    moviendo = IZQUIERDA
                elif evento.key == pygame.K_RIGHT:
                    moviendo = DERECHA
                elif evento.key == pygame.K_g:
                    if estadoBalon == INVISIBLE:
                        estadoBalon = APARECE

                elif evento.key == pygame.K_SPACE:
                    if estado==GANAR or estado==PERDER:
                        efecto.play()
                        efecto3.stop()
                        efecto4.stop()
                        listaTargets = []
                        estado = JUGANDO
                        efecto.play()

                elif evento.key == pygame.K_ESCAPE:
                    ARCHIVO.close()
                    termina = True


            elif evento.type == pygame.MOUSEBUTTONUP:
                if estado==MENU:
                    xm, ym = pygame.mouse.get_pos()
                    xb = 80
                    yb = +400
                    if xm>=xb and xm<=xb+450 and ym >= yb and ym <=yb+100:
                        efecto.play(0)
                        pygame.mixer.music.stop()
                        estado = JUGANDO
                        

        ventana.fill(BLANCO)

        if estado == JUGANDO:
            ventana.blit(imgFondo, (0,0))
            dibujarPersonaje(ventana, spritePersonaje)
            dibujarEnemigos(ventana, listaEnemigos)
            dibujarEnemigos(ventana, listaLays)
            dibujarBalones(ventana, listaBalones)
            moverBalones(listaBalones)
            verificarColision(listaEnemigos,listaBalones,listaTargets)
            verificarColision(listaLays,listaBalones,listaTargetsMalos)
            verificarSlida(listaBalones)
            moverBalones(listaBalones)
            if moviendo == ARRIBA and spritePersonaje.rect.bottom>125:
                spritePersonaje.rect.bottom -=5
            elif moviendo == ABAJO and spritePersonaje.rect.bottom<ALTO:
                spritePersonaje.rect.bottom += 5
            elif moviendo == IZQUIERDA and spritePersonaje.rect.left>0:
                spritePersonaje.rect.left -= 5
            elif moviendo == DERECHA and spritePersonaje.rect.left<736:
                spritePersonaje.rect.left += 5

            if estadoBalon == APARECE:
                efecto2.play(0)
                spriteBala = pygame.sprite.Sprite()
                spriteBala.image = imgBalon
                spriteBala.rect = imgBalon.get_rect()
                spriteBala.rect.left = spritePersonaje.rect.left + spritePersonaje.rect.width
                spriteBala.rect.bottom = spritePersonaje.rect.bottom
                listaBalones.append(spriteBala)
                estadoBalon=VIAJANDO

            texto = fuente.render("Tiempo: %d" % (timer), 1, NEGRO)
            ventana.blit(texto, ( ANCHO//2+ 300, ALTO-70))
            texto6 = fuente6.render("Score: %d" % (len(listaTargets)), 1, NEGRO)
            ventana.blit(texto6, ( ANCHO//2+ 300, ALTO -50))
            texto2= fuente6.render('Move: < ^ > Shoot: G', 1, NEGRO)
            ventana.blit(texto2, ( 20,ALTO-50))
            facto=factos[randint(0,7)]

            if len(listaTargetsMalos) > 1:
                efecto4.play()
                estado=PERDER

        if estado == PERDER:
            ventana.blit(img2, (0, 0))
            texto2 = fuente2.render("Press space to continue", 1, ROJO)
            ventana.blit(texto2, ( 0, ALTO-80))
            
            ventana.blit(imgLOSE, (250, 50))


            #Imprime el facto
            textoFacto = fuente5.render(facto, 1, ROJO)

            char_width = fuente5.size('A')[0]
            chars_per_line = ANCHO // char_width - 40

            # Divide el texto en líneas
            lines = textwrap.wrap(facto, width=chars_per_line)

            # Imprime cada línea
            for i, line in enumerate(lines):
                text = fuente5.render(line, True, BLANCO)
                ventana.blit(text, (30, 200+i*25))  # Asume que quieres un espacio de 25 píxeles entre líneas


            # ventana.blit(textoFacto, (30, ALTO // 2))

            texto4 = fuente4.render("Score: %d" % (len(listaTargets)), 1, ROJO)
            ventana.blit(texto4, (ANCHO - 200, 0))
            listaEnemigos =[]
            listaBalones=[]
            listaTargetsMalos=[]
            listaLays=[]
            timer = 0

            # Personaje
            spritePersonaje = pygame.sprite.Sprite()
            spritePersonaje.image = imgPersonaje
            spritePersonaje.rect = imgPersonaje.get_rect()
            spritePersonaje.rect.left = 0

            imgEnemigo = pygame.image.load("pepsi_zero.png")
            imgEnemigo= pygame.transform.scale(imgEnemigo, (60, 60))

            for k in range(6):
                spriteEnemigo = pygame.sprite.Sprite()
                spriteEnemigo.image = imgEnemigo
                spriteEnemigo.rect = imgEnemigo.get_rect()
                spriteEnemigo.rect.left = randint(200, ANCHO - 100)
                spriteEnemigo.rect.bottom = randint(spriteEnemigo.rect.height + 50, ALTO - 10)
                listaEnemigos.append(spriteEnemigo)
            
            for k in range(randint(1,4)):
                spriteLays = pygame.sprite.Sprite()
                spriteLays.image = imgLays
                spriteLays.rect = imgLays.get_rect()
                spriteLays.rect.left = randint(200, ANCHO-100)
                spriteLays.rect.bottom = randint(spriteLays.rect.height+50, ALTO-10)
                listaLays.append(spriteLays)

                spriteLays = pygame.sprite.Sprite()
                spriteLays.image = imgCheetos
                spriteLays.rect = imgCheetos.get_rect()
                spriteLays.rect.left = randint(200, ANCHO-100)
                spriteLays.rect.bottom = randint(spriteLays.rect.height+50, ALTO-10)
                listaLays.append(spriteLays)

                spriteLays = pygame.sprite.Sprite()
                spriteLays.image = imgdoritos
                spriteLays.rect = imgdoritos.get_rect()
                spriteLays.rect.left = randint(200, ANCHO-100)
                spriteLays.rect.bottom = randint(spriteLays.rect.height+50, ALTO-10)
                listaLays.append(spriteLays)

        if len(listaTargets) == 5:
            efecto3.play()
            estado = GANAR
            if timer < record and timer>10:
                ARCHIVO.write ("\n"+str(timer))
                timer=0

        if estado == GANAR:
            ventana.blit(img3, (0, 0))
            listaEnemigos =[]
            listaTargetsMalos=[]
            listaLays=[]
            listaBalones=[]
            timer = 0
            texto3 = fuente3.render("Press space to play again", 1, NEGRO)
            ventana.blit(texto3, (190, 100))
            ventana.blit(imgWIN, (250, +400))
            char_width = fuente5.size('A')[0]
            chars_per_line = ANCHO // char_width - 40

            # Divide el texto en líneas
            lines = textwrap.wrap(facto, width=chars_per_line)

            # Imprime cada línea
            for i, line in enumerate(lines):
                text = fuente5.render(line, True, BLANCO)
                ventana.blit(text, (30, 200+i*25))  # Asume que quieres un espacio de 25 píxeles entre líneas

            # Personaje
            spritePersonaje = pygame.sprite.Sprite()
            spritePersonaje.image = imgPersonaje
            spritePersonaje.rect = imgPersonaje.get_rect()
            spritePersonaje.rect.left = 0

            imgEnemigo = pygame.image.load("pepsi_zero.png")    
            imgEnemigo= pygame.transform.scale(imgEnemigo, (60, 60))
            for k in range(6):
                spriteEnemigo = pygame.sprite.Sprite()
                spriteEnemigo.image = imgEnemigo
                spriteEnemigo.rect = imgEnemigo.get_rect()
                spriteEnemigo.rect.left = randint(200, ANCHO - 100)
                spriteEnemigo.rect.bottom = randint(spriteEnemigo.rect.height + 50, ALTO - 10)
                listaEnemigos.append(spriteEnemigo)
            for k in range(randint(1,4)):
                spriteLays = pygame.sprite.Sprite()
                spriteLays.image = imgLays
                spriteLays.rect = imgLays.get_rect()
                spriteLays.rect.left = randint(200, ANCHO-100)
                spriteLays.rect.bottom = randint(spriteLays.rect.height+50, ALTO-10)
                listaLays.append(spriteLays)

                spriteLays = pygame.sprite.Sprite()
                spriteLays.image = imgCheetos
                spriteLays.rect = imgCheetos.get_rect()
                spriteLays.rect.left = randint(200, ANCHO-100)
                spriteLays.rect.bottom = randint(spriteLays.rect.height+50, ALTO-10)
                listaLays.append(spriteLays)

                spriteLays = pygame.sprite.Sprite()
                spriteLays.image = imgdoritos
                spriteLays.rect = imgdoritos.get_rect()
                spriteLays.rect.left = randint(200, ANCHO-100)
                spriteLays.rect.bottom = randint(spriteLays.rect.height+50, ALTO-10)
                listaLays.append(spriteLays)

        if estado == MENU:
            timer = 0
            ventana.blit(imgFondoMenu, (0, 0))
            ventana.blit(imgRuffle, (170, 20))
            
            textoMenu = fuente3.render("Help Ruffilio practice his passing.", 1, BLANCO)
            ventana.blit(textoMenu, (10,ALTO//2-80))
            textoMenu2 = fuente3.render("Hit all the Pepsi cans avoiding", 1, BLANCO)
            ventana.blit(textoMenu2, (10,ALTO//2-40))
            textoMenu2 = fuente3.render("the Lay´sCheetos and Doritos bags.", 1, BLANCO)
            ventana.blit(textoMenu2, (10,ALTO//2))

            dibujarMenu(ventana, imgBtnJugar)

        pygame.display.flip()
        reloj.tick(20)
        timer += 1/40
    ARCHIVO.close()
    pygame.quit()

def main():
    dibujar()

main()
```
